```python
df1= pd.read_csv('https://raw.githubusercontent.com/ayyucedemirbas/machine_learning_algorithms/master/datasets/world-happiness-report.csv')
```

```python
f, axes = plt.subplots(5, 2, figsize=(20,30))
f.tight_layout(pad=8)
f.suptitle("Histogram Plots")
cols=df1.select_dtypes(exclude='object').columns

x_axes=0
y_axes=0

for col in cols:
  sns.histplot(data=df1, x=col, kde=True, ax=axes[x_axes, y_axes])
  if y_axes == 1:
    y_axes=0
    x_axes+=1
  else:
    y_axes +=1

plt.show()
```

```python
f, axes = plt.subplots(5, 2, figsize=(20,30))
f.tight_layout(pad=8)
f.suptitle("Box Plots")
cols=df1.select_dtypes(exclude='object').columns

x_axes=0
y_axes=0

for col in cols:
  sns.boxplot(data=df1, x=col, ax=axes[x_axes, y_axes])
  if y_axes == 1:
    y_axes=0
    x_axes+=1
  else:
    y_axes +=1

plt.show()
```
