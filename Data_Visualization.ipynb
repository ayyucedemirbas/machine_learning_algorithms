```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x= np.linspace(0,10,100)
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')


plt.legend(('sin', 'cos'))
```

```python
fig = plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x), '-')

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), '--', linewidth = 4, color='red')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')

plt.legend(('cos', 'cos'))
```

```python
fig.savefig('data.png')
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_length']
```

```python
iris['petal_length'].plot()
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index)
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index, hue= 'species')
```

```python
sns.lineplot(data=iris, y='sepal_length', x=iris.index, hue= 'species')
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
iris.groupby(by='species').mean()
```

```python
iris.groupby(by='species').max()
```

```python
iris.groupby(by='species').sum()
```

```python
iris.groupby(by='species').mean()['sepal_length']
```

```python
data = iris.groupby(by='species').mean()['sepal_length']
ax = sns.barplot(x=data.index, y=data.values)
ax.set_title('Sepal Length Means')
ax.set_ylabel('Mean')
```

```python
sns.histplot(data=iris, x= 'petal_length')
```

```python
sns.histplot(data=iris, x= 'sepal_width')
```

```python
sns.histplot(data=iris, y= 'sepal_width')
```

```python
sns.histplot(data=iris, x= 'sepal_width', hue='species')
```

```python
sns.histplot(data=iris, x= 'sepal_width', bins=20)
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.boxplot(data=iris, y='sepal_length', x= 'species')
```
